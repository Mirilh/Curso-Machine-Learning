### Tarea 4: Crea una columna Class utilizando LandingOutcome


Previamente hemos guardado en `landing_outcomes` el recuento de los diferentes aterrizajes:

```python
print(landing_outcomes)
```

*La variable anterior es un diccionario, utilizandola crea un conjunto con los diferentes casos donde el cohete no aterrizó satisfactoriamente. Llama a este conjunto `bad_outcomes`*

```python
#SOLUCION: {'False ASDS', 'False Ocean', 'False RTLS', 'None ASDS', 'None None'}
lista = [] #lista temporal para pasar a conjunto y que elimine los valores repetidos

for index, value in landing_outcomes.items(): #se recorre como un diccionario 
  if index != 'True':
    # print(index)
    lista.append(index)

bad_outcomes = set(lista) #funcion set para pasar de lista a conjunto
print(bad_outcomes)
```

Creemonos ahora una lista llamada `landing_class` que tomará valores 0 y 1 según si el resultado del aterrizaje satisfactorio o no. (True 1, otros 0) del data frame original que es el que tiene todos los datos.
Pasos:
* *Recorre para cada celda la columna `LandingOutcome`.*
* *Comprueba si el valor pertenece al conjunto `bad_outcomes`, es decir tiene True o si no.  adjunta un 0 a la lista `landing_class` si no tiene true, si no, adjunta un 1.*

```python
landing_class = []

for value in df.LandingOutcome: #se recorre sobre el original
  if 'True' in value:
    landing_class.append(1)
  else:
    landing_class.append(0)
print(landing_class)
```

*Añade esta lista al dataframe como una columna con nombre `Class`*

```python
df['Class'] = landing_class

df.head()
```
