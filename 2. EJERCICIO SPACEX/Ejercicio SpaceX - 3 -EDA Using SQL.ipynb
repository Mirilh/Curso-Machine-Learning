##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query = """SELECT LaunchSite	 
FROM SPACEXTBL 
WHERE LaunchSite like '%CCS%' 
LIMIT 5 ;""" 

df = pd.read_sql(query, con = engine )
df
```

## Ejercicio 3

##### Muestra la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """SELECT 
            sum(CASE 
                  WHEN Customers like '%NASA (CRS)%' THEN PayloadMass
                  END)AS suma_NASA
          FROM SPACEXTBL;""" 

df = pd.read_sql(query, con = engine )
df
```

```python
query = """SELECT 
            sum(PayloadMass) AS suma_NASA
          FROM SPACEXTBL
          WHERE Customers like '%NASA (CRS)%'
          ;""" 

df = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """SELECT 
            AVG(PayloadMass) AS media_NASA
          FROM SPACEXTBL
          WHERE Serial like 'B1%'
          ;""" 

df = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """SELECT
            Date
          FROM SPACEXTBL
          WHERE LandingOutcome LIKE '%True RTLS%'
          GROUP BY DATE 
          LIMIT 1 ;""" 

df = pd.read_sql(query, con = engine )
df
```
